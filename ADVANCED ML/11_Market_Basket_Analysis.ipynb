```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Install openpyxl package
%pip install openpyxl

# Read the Excel file
df = pd.read_excel(r'Online_Retail.xlsx')
df.sample(20)
```

```python
# Removing spaces from the description
df['Description'] = df['Description'].str.strip()

# Droping rows without invoice number
df.dropna(axis=0, subset=['InvoiceNo'], inplace=True)
df['InvoiceNo'] = df['InvoiceNo'].astype('str')

# Removing the credit card transaction with invoice numbers containing C
df = df[~df['InvoiceNo'].str.contains('C')]
df
```

```python
basket=(df[df['Country']=='France']
          .groupby(['InvoiceNo', 'Description'])['Quantity']
          .sum()
            .unstack()
            .reset_index()
            .fillna(0)
          .set_index('InvoiceNo'))
basket
```

```python
def encode_units(x):
      if x <= 0:
         return 0
      if x >= 1:
         return 1
      
basket_sets = basket.applymap(encode_units)
basket_sets.drop('POSTAGE', inplace=True, axis=1)
basket_sets.head()
```

```python
# Generating frequent item sets that have a support of at least 7%
frequent_itemsets = apriori(basket_sets, min_support=0.09, use_colnames=True)

# Generating the rules with their corresponding support, confidence and lift values 
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

rules.head()
```

```python
# Generating frequent item sets that have a support of at least 7%
frequent_itemsets = apriori(basket_sets, min_support=0.07, use_colnames=True)

# Generating the rules with their corresponding support, confidence and lift values 
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)

rules.head()
```

```python
# Filtering out the values with lift > = 6 and confidence > = 0.8
rules[ (rules['lift'] >= 6) &
       (rules['confidence'] >= 0.8) ]
```
