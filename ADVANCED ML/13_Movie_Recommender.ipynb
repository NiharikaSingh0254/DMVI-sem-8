```python
import pandas as pd
import numpy as np
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head()
# budget
# homepage
# id
# original_language
# original_title
# popularity
# production_comapny
# production_countries
# release-date(not sure)
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
# preprocessing genres

def convert(text):
      L=[]
      for i in ast.literal_eval(text):
            L.append(i['name'])
      return L
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres']=movies['genres'].apply(convert)

movies.head()
```

```python
# Preprocessing keywords column

movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# preprocessing cast column

def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
# Preprocessing crew column fetch the director

def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
        return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# Movie overview column convert to list

movies['overview'][0]
```

```python
# Type 
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# Remove the white space of the text like a sam worthing to convert SamWorthington
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x] if x is not None else [])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x] if x is not None else [])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x] if x is not None else [])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x] if x is not None else [])
```

```python
movies.head()
```

```python
# Create a new column tag to concate all column (genres,overview,keywords,crew)
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['cast']  + movies['keywords'] + movies['crew']
```

```python
movies.head()
```

```python
# Create new DataFrame

new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
# tags convert into strings

new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
# string convert in lower case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
# Text Vectorization (Text -> Vector = Vectorization) Usibg Bag of Words 
# All tags concate and than most find the most repeatable words in tags
# Word1,word2,word3........words50000 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')

```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names_out()
```

```python
# Distance Between two vector
# It will calculate distance between another vector


from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

```python
# List to tupple data convert sorting all data in accendinng order

sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])
```

```python
# Find idex number of any perticular movie

new_df[new_df['title'] == 'Batman Begins'].index[0]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Batman Begins')
```
