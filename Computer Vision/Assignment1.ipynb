# **1.  Load the image in grayscale**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Loading image and converting to GRAYSCALE
image_path = 'img1.jpg'  
image_gray = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)


# Display image
plt.figure(figsize=(10, 5))

# Original Image
image_color = cv2.imread(image_path) 
image_color = cv2.cvtColor(image_color, cv2.COLOR_BGR2RGB)  

plt.subplot(1, 2, 1)
plt.imshow(image_color)
plt.title('Original Image')
plt.axis('off')

# Grayscale Image
plt.subplot(1, 2, 2)
plt.imshow(image_gray, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.show()
```

# **2. Histogram Equalization**

```python
image_eq = cv2.equalizeHist(image_gray)
plt.subplot(1, 1,1)
plt.imshow(image_eq, cmap='gray')
plt.title('Histogram Equalized Image')
plt.axis('off')
plt.show()
```

# **3. Smoothing using Gaussian and Median Filters**

```python
image_gaussian = cv2.GaussianBlur(image_gray, (5, 5), 0)
image_median = cv2.medianBlur(image_gray, 5)

# Original Image
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(image_gray, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Gaussian Blurred Image
plt.subplot(1, 3, 2)
plt.imshow(image_gaussian, cmap='gray')
plt.title('Gaussian Blurred')
plt.axis('off')

# Median Blurred Image
plt.subplot(1, 3, 3)
plt.imshow(image_median, cmap='gray')
plt.title('Median Blurred')
plt.axis('off')
plt.show()
```

# **4. Sharpening using Kernel**

```python
sharpening_kernel = np.array([[0, -1, 0],
                               [-1, 5, -1],
                               [0, -1, 0]])
image_sharp = cv2.filter2D(image_gray, -1, sharpening_kernel)

# Original Image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image_gray, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Sharpened Image
plt.subplot(1, 2, 2)
plt.imshow(image_sharp, cmap='gray')
plt.title('Sharpened Image')
plt.axis('off')
plt.show()
```

# **5. Edge Detection using Canny**

```python
edges = cv2.Canny(image_gray, 100, 200)

# Original Image
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image_gray, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Canny Edge Detection 
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Canny Edge Detection')
plt.axis('off')
plt.show()
```
