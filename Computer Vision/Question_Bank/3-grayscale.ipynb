### **Write an OpenCV python script that takes a color image as an input, and converts it to grayscale. Access the pixel value at [x, y] (i.e., [100.150]) location and change the value to a different value. Display the original image and image after changing the pixel values side by side.** 

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display images side by side
def show_images(images, titles, cmap=None):
    plt.figure(figsize=(10,5))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, len(images), i+1)
        plt.imshow(img, cmap=cmap)
        plt.title(title)
        plt.axis('off')
    plt.show()

# Load a color image
image = cv2.imread('..\CV-1.jpg')  # Replace with your image path
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Modify a specific pixel value
x, y = 100, 150  # Example coordinates
gray[y, x] = 200  # Change pixel value

# Display Original and Modified Grayscale images
show_images([image_rgb, gray], ['Original Image', 'Modified Grayscale Image'], cmap='gray')
```
