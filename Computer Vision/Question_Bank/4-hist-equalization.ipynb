### **Write an OpenCV Python script that performs histogram equalization on a given image. Display original and equalized images.**

```python
# Import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Function to display multiple images
def show_images(images, titles, cmap=None):
    plt.figure(figsize=(15,5))
    for i, (img, title) in enumerate(zip(images, titles)):
        plt.subplot(1, len(images), i+1)
        plt.imshow(img, cmap=cmap)
        plt.title(title)
        plt.axis('off')
    plt.show()

# Load a color image using OpenCV
image = cv2.imread('..\CV-1.jpg')  # Replace 'image.jpg' with the actual image path
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Convert to Grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Modify a specific pixel value
x, y = 100, 150  # Example coordinates
gray[y, x] = 200  # Change pixel value to a new intensity

# Apply Histogram Equalization
equalized = cv2.equalizeHist(gray)

# Display Original, Modified, and Equalized Grayscale images
show_images([image_rgb, gray, equalized], ['Original Image', 'Modified Grayscale Image', 'Histogram Equalized'], cmap='gray')
```
