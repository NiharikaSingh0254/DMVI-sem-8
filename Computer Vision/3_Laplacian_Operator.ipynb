### **Laplacian Operator**

```python
import cv2
from matplotlib import pyplot as plt
```

```python
# Load the image
img = cv2.imread('cv4.jpg')

# Convert the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# Apply Laplacian operator
laplacian = cv2.Laplacian(gray, cv2.CV_64F)

# Sharpened image
sharp = cv2.convertScaleAbs(laplacian)

# Display the results
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(gray,cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Sharpened Image")
plt.imshow(sharp,cmap='gray')

plt.show()
```

### **Laplacian Operator for colored image**

```python
# load the image
img = cv2.imread('cv5.jpg')

# Convert the image to grayscale
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Apply Laplacian operator
laplacian = cv2.Laplacian(img, cv2.CV_64F)
laplacian = cv2.convertScaleAbs(laplacian)

# Add the Laplacian result back to original image for sharpening
sharp=cv2.addWeighted(img,1.5,laplacian,-0.5,0)

# Display the result
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img)

plt.subplot(1, 2, 2)
plt.title("Sharpened Image")
plt.imshow(sharp)
plt.show()
```

```python
# load the image
img = cv2.imread('cv6.jpg')

# Convert the image to grayscale
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Apply Laplacian operator
laplacian = cv2.Laplacian(img, cv2.CV_64F,ksize=5)
laplacian = cv2.convertScaleAbs(laplacian)

# Add the Laplacian result back to original image for sharpening
sharp=cv2.addWeighted(img,1.5,laplacian,-0.5,0)

# Display the result
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(img)

plt.subplot(1, 2, 2)
plt.title("Sharpened Image")
plt.imshow(sharp)
plt.show()
```
